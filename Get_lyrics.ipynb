```python
#https://lyricsgenius.readthedocs.io/en/master/reference/types.html#lyricsgenius.types.Song
#https://github.com/johnwmillr/LyricsGenius/blob/master/docs/src/index.rst#id1
from lyricsgenius import Genius
import pandas as pd
from tqdm.auto import tqdm
```

```python
df = pd.read_csv('dataset.csv', dtype = str)
df = df.drop_duplicates('track_id').reset_index(drop = True)
```

```python
df = df[['track_name','artists']]
df['lyrics'] = ['' for i in range(len(df))]
df['language'] = ['' for i in range(len(df))]
df['lyrics_state'] = ['' for i in range(len(df))]
df['Song_id'] = ['' for i in range(len(df))]
df['url'] = ['' for i in range(len(df))]
df['release_date'] = ['' for i in range(len(df))]
#df.to_csv('dataset_with_lyrics.csv',index = False)

token = '2enXjSvGsvlu7juxcR5m9hsXe_3XHSkUmJ4BKQ-ShGsuTbLhnv7HFUXjBw43nizB'
genius = Genius(token)

for i in tqdm(range(22500)):
    song_name = df['track_name'][i]
    artist = df['artists'][i]
    try: 
        song = genius.search_song(song_name,artist = artist).to_dict()
        df['lyrics'][i] = song['lyrics']
        df['lyrics_state'][i] = song['lyrics_state']
        df['language'][i] = song['language']
        df['Song_id'][i] = str(song['id'])
        df['url'][i] = song['url']
        df['release_date'][i] = song['release_date_for_display']
    except:
        df['lyrics'][i] = 'NOT FOUND'
        print(f'{song_name} not found')
    
    #prevent broken, save progress every 100 songs
    if (i+1)%200 == 0:
        df.to_csv('dataset_with_lyrics.csv',index = False)
        print(f'-------{i+1} songs has been saved to file-------')
```
